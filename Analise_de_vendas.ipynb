```python
df_revenue = df.pivot_table(values='Revenue', index=['Retailer country'],
                            aggfunc=('sum', 'count', 'mean'))
```

```python
df_revenue
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(x='sum', data=df_revenue);
```

```python
nome_paises = df_revenue.index
```

```python
plt.pie(x='sum', labels= nome_paises ,data=df_revenue);
```

```python
sns.set(rc={'figure.figsize' : (15 , 12)})
plt.pie(x='sum', labels= nome_paises , autopct= '%.2f%%', pctdistance=0.8 ,  shadow=True ,data=df_revenue);
```
