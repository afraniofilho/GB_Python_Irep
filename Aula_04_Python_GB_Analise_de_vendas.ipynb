```python
# Vamos criar uma variavel (df_revenue), total de vendas (Count), média da receita (mean), somatório de receita
# por país
```

```python
df_revenue = df.pivot_table(values='Revenue', index=['Retailer country'], aggfunc=('sum', 'count', 'mean'))
```

```python
df_revenue
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(x='sum', data=df_revenue);
```

```python
nome_paises = df_revenue.index
```

```python
nome_paises
```

```python
plt.pie(x='sum', labels= nome_paises, data=df_revenue)
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
plt.pie(x='sum', labels= nome_paises, data=df_revenue);
```

```python
plt.pie(x='sum', labels= nome_paises, labeldistance=1.2, autopct='%.2f%%' , pctdistance= 0.8,shadow=True, data=df_revenue);
```
