```
import pandas as pd
import seaborn as sns
```

```
df = pd.read_csv('Sales_Products.csv')
```

```
df.head()
```

```
df.shape
```

```
## Qual a quantidade de vendas por país?
```

```
df_country = df['Retailer country'].value_counts()
```

```
df_country
```

```
df_country.plot.bar(width=0.4, color='red')
```

```
## Quais os somatórios da receita (Revenue) das vendas por país?
```

```
df_revenue = df.pivot_table(values='Revenue', index=['Retailer country'],
                            aggfunc=('sum', 'count', 'mean'))
```

```
df_revenue
```

```
import matplotlib.pyplot as plt
```

```
plt.pie(x='sum', data=df_revenue)
```

```
nome_paises = df_revenue.index.values
```

```
nome_paises
```

```
sns.set(rc={'figure.figsize':(15,12)})
plt.pie(x='sum', data=df_revenue, labels=nome_paises, autopct='%.2f%%',pctdistance=0.9,rotatelabels=True)
```

```
w,l,p = plt.pie(x='sum', data=df_revenue, labels=nome_paises, autopct='%.2f%%',pctdistance=0.9,rotatelabels=True)
[t.set_rotation(60) for t in p]
```
